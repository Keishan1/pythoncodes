```python
#Import all the necessary libraries 
import requests
from bs4 import BeautifulSoup

import numpy as np

import pandas as pd
```

```python
#Website being scraped, and making request to
URL = 'https://www.iqair.com/world-most-polluted-countries'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#List each data will be appened to
Rank = []
Country_Region = []
AVG_2020 = []
AVG_2019 = []
AVG_2018 = []
Population_2021 = []

#find the table
table = soup.find('table')

#for each row in table find all the tr tags and then all the td tags within them 
#and if the length if the table has 6 columns 
for row in table.find_all('tr'):
  cells = row.findAll('td')
  if(len(cells) == 6):
    
    #append to the Rank list if the td tag was in the 1st column
    Rank.append(cells[0].get_text())
    
    #append to the Country list if the td tag was in the 2nd column
    Country_Region.append(cells[1].get_text())
    
    #append to the Avg_2020 list if the td tag was in the 3rd column,
    #and turn it into  float
    AVG_2020.append(float(cells[2].get_text()))
    
    #In the 4th colomn in the table append all the nan/ blank cells
    #this way I can use the float method on the non blank cells 
    if len(cells[3].get_text()) == 3:
      AVG_2019.append(np.nan)
    else:
      AVG_2019.append(float(cells[3].get_text()))

    #In the 5th colomn in the table append all the nan/ blank cells
    #this way I can use the float method on the non blank cells 
    if len(cells[4].get_text()) == 3:
      AVG_2018.append(np.nan)
    else:
      AVG_2018.append(float(cells[4].get_text()))

    #In the 6th column in the table append all the cells and then 
    #remove the commas from the cells so I can turn the string intan int
    Pop = (cells[5].get_text())
    Pop2 = Pop.replace(',','')
    Population_2021.append(int(Pop2))
    
```

```python
#To create the data frame
df = pd.DataFrame()

#To name each column and state the list the column will get its numbers from 
df["Rank"] = Rank
df["Country Region"] = Country_Region
df["AVG 2020"] = AVG_2020
df["AVG 2019"] = AVG_2019
df["AVG 2018"] = AVG_2018
df["Population 2021"] = Population_2021

```
