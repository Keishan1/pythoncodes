```python
#Import all the necessary libraries 
import requests
import json

import pandas as pd
```

```python
#List each data will be appened to
Name = []
Temp = []
Humidity = []
Weather_des = []
Max_Temp = []
Min_Temp = []
Wind_speed = []

#api Key
key = 'effa8ae07840dcaffea3c9b3cb3380c8'

#cities making request for
cities = ['Tokyo', 'Lagos', 'Istanbul', 'Mexico City', 'Sao Paulo', 'Sydney', 'McMurdo Station', 'Montego Bay', 'Oymyakon', ' Mecca']

#for each city in the cities list above make a request from the specified website
#if the request code is 200 meaning the request was sucessful, and append each
#item to the corresponding list
for city in cities:
  url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={key}&units=imperial'
  r = requests.get(url)
  if(r.status_code == 200):
    data = r.json()
    Name.append(data['name'])
    Temp.append(data['main']['temp'])
    Humidity.append(data['main']['humidity'])
    Weather_des.append(data['weather'][0]['description'])
    Max_Temp.append(data['main']['temp_max'])
    Min_Temp.append(data['main']['temp_min'])
    Wind_speed.append(data['wind']['speed'])
  

    
```

```python
#created a data frame using the appended data 
DF = pd.DataFrame()

DF['Name'] = Name
DF['Temp'] = Temp
DF['Humidity'] = Humidity
DF['Weather Description'] = Weather_des
DF['Max Temp'] = Max_Temp
DF['Min Temp'] = Min_Temp
DF['Wind Speed'] = Wind_speed
```

```python
#generats a csv file, with data, 
DF.to_csv('Weather.csv')
```
