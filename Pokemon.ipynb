```python
#Import all the necessary libraries 
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
#Website being scraped, and making request to
URL = 'https://pokemondb.net/pokedex/all'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#find table on webpage
table = soup.find('table')
```

```python
#List each data will be appened to
IDS= []
Name = []
Type = []
Total = []
HP = []

#for all the column in the table append them to there corresponding list,
#int is stated to turn certain columns into intergers to be
#able to describe method
for row in table.find_all('tr'):
  cells = row.findAll('td')
  if(len(cells) == 10):
    IDS.append(int(cells[0].get_text()))
    Name.append(cells[1].get_text())
    Type.append(cells[2].get_text())
    Total.append(int(cells[3].get_text()))
    HP.append(int(cells[4].get_text()))  
```

```python
#To create the data frame
DF1 = pd.DataFrame()

#To name each column and state the list the colum will get its numbers from 
DF1['ID'] = IDS
DF1['Name'] = Name
DF1['Type'] = Type
DF1['Total'] = Total
DF1['HP'] = HP
```

```python
#import json library
import requests
import json
```

```python
#List each data will be appened to
Attack = []
Defense = []
Sp_attack = []
Sp_defense = []
Speed = []

#for each item in the list called ID a request will be made to this API Key
#that will return and append the specified data, if and only if thet staus code is
#200 when request is made
for ID in IDS:
  url = f"https://pokeapi.co/api/v2/pokemon/{ID}/"
  r = requests.get(url)
  data = r.json()
  if(r.status_code == 200):
    Attack.append(int(data['stats'][1]['base_stat']))
    Defense.append(int(data['stats'][2]['base_stat']))
    Sp_attack.append(int(data['stats'][3]['base_stat']))
    Sp_defense.append(int(data['stats'][4]['base_stat']))
    Speed.append(int(data['stats'][5]['base_stat']))
  
```

```python
#To create the data frame
DF2 = pd.DataFrame()

#To name each column and state the list the column will get its numbers from 
DF2['Attack'] = Attack
DF2['Defense'] = Defense
DF2['Sp. Attack'] = Sp_attack
DF2['Sp. Defense'] = Sp_defense
DF2['Speed'] = Speed
```

```python
DF2
```

```python
#used concat to join the two data frames togther 
DF3 = pd.concat([DF1, DF2], axis=1)
```

```python
#generats a csv file, with data
DF3.to_csv('Pokemon.csv')
```
