```python
#import the requests libraries
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
#defined the function definition to take parameter 'URLS'
def find_heights (URLS):
  all_df = []
  for URL in URLS:
    page = requests.get(URL)

    soup = BeautifulSoup(page.content, 'html.parser')

    #find all td tags with the class name equal to side..., this was where the player 
    #names were located
    names = soup.find_all('td', {'class':'sidearm-table-player-name'})

    #find all td tags with the class name equal to height
    heights = soup.find_all('td', {'class':'height'})
    

    player_heights = []
    player_names = []

    #inspect every value thats in the variable table & 
    #assign it to the variable feet if when sliced it's the 1st index in that item 
    #then turn it into a integer and multiply it by 12 since 1ft = 12inches
    #other wise assign it to the variable inches if when sliced it's the 3rd index or 
    #anything after in that item, in this case there's nothing after 
    #then turn it into a integer and multiply 
    #Next centimeters becomes what you get when feet & inches are added 
    #so create a new list called team_heights that will be a list of all the variables of centimeters
    for heights in heights:
      feet = (int(heights.get_text()[0:1])) * 12
      inches = (int(heights.get_text()[2:])) 
      total_inches = feet + inches 
      player_heights.append(total_inches)
  
  #Did the same thing I did above for heights
    for names in names: 
      get_names = (names.get_text()[1:-1]) 
      player_names.append(get_names)

  #Created a dataframe the key here is the colomn name which is "Names","Heights" and 
  #the value in the list are player_names and player_heights
    Team = {'Name': player_names, 'Height': player_heights}  

   #created a pandas dataframe from Team dataframe  
    df = pd.DataFrame(Team)

  #since this function is built to take multiple URLS, this is will append each df
    all_df.append(df)

  #combine all the dataframes from the URLS in each specific list 
    All_combined = pd.concat(all_df, ignore_index = True)

  #sort each combined dataframe by from the shortest to the tallest players 
    All_combined.sort_values(by='Height', inplace = True)

  #what is returned after all the fuctions above are passed throught the dataframes
  return All_combined
```

```python
#Created a list with the websites since I made a function that can be called for multiple URLS

Men_VB_websites = ['https://www.brooklyncollegeathletics.com/sports/mens-volleyball/roster/2019',
                   'https://athletics.baruch.cuny.edu/sports/mens-volleyball/roster',
                   'https://yorkathletics.com/sports/mens-volleyball/roster']

Women_VB_websites = ['https://www.brooklyncollegeathletics.com/sports/womens-volleyball/roster/2019',
                     'https://athletics.baruch.cuny.edu/sports/womens-volleyball/roster',
                     'https://johnjayathletics.com/sports/womens-volleyball/roster']

Men_S_websites = ['https://www.brooklyncollegeathletics.com/sports/mens-swimming-and-diving/roster',
                   'https://athletics.baruch.cuny.edu/sports/mens-swimming-and-diving/roster',
                   'https://yorkathletics.com/sports/mens-swimming-and-diving/roster']

Women_S_websites = ['https://www.brooklyncollegeathletics.com/sports/womens-swimming-and-diving/roster',
                     'https://athletics.baruch.cuny.edu/sports/mens-swimming-and-diving/roster',
                    'https://queensknights.com/sports/womens-swimming-and-diving/roster/2019-20']
```

```python
#Assigned variables to each list value above so later on I can call the method
#for each variable since it doesn't work with list items, I have to convert it
#to an actual variable first, also so I can concat the dataframes

Men_VB_df = find_heights(Men_VB_websites)

Women_VB_df = find_heights(Women_VB_websites)
 
Men_S_df = find_heights(Men_S_websites)

Women_S_df = find_heights(Women_S_websites)

#Created a list to be able to give each dataframe a header
keys = ["Men's Volleyball Team","Men's Swim Team","Women's Volleyball Team","Women's Swim Team"]
All_combined =pd.concat([Men_VB_df, Men_S_df,Women_VB_df, Women_S_df ], axis = 1, keys = keys)


```

```python
print('Listed below are all the players for each team/sport:')
All_combined
```
